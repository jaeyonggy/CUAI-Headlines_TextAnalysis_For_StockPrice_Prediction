#### 주식 데이터에서 해당 주식의 날짜에 맞는 주가변동여부와 거래량변동여부를 기사 제목 데이터에 넣으려고 한다

```python
data_stock = pd.read_csv('data_stock.csv')
data_title = pd.read_csv('data_title.csv')
```

### 관심 주식

```python
stocks = ['한전기술','셀트리온','카카오게임즈','iMBC','하이브',
          '삼성전자','현대자동차','HMM','대한항공','NAVER','두산중공업','SK하이닉스']  
# 한국 거래소에 현대차가 현대자동차로, 네이버가 NAVER로, sk하이닉스가 SK하이닉스로 나와있다.
```

#### data_title의 주식 이름을 data_stock의 쓰여진걸로 바꿔줘야 할듯

```python
data_title['주식'] = data_title['주식'].apply(lambda x:x.replace('현대차', '현대자동차'))
data_title['주식'] = data_title['주식'].apply(lambda x:x.replace('네이버', 'NAVER'))
data_title['주식'] = data_title['주식'].apply(lambda x:x.replace('sk하이닉스', 'SK하이닉스'))
```

```python
data_title
```

### 날짜를 시간타입으로 다루자

```python
from datetime import datetime, timedelta  # timedelta는 시간타입 자료에 하루를 더하기용

data_title['날짜'] = pd.to_datetime(data_title['날짜'])
data_stock['날짜'] = pd.to_datetime(data_stock['날짜'])
```

### 주식 장은 평일에만 열린다. 따라서 기사가 쓰인 다음 날이 휴일이나 주말이여서 장이 안 열린 경우 해당 날짜의 기사들을 지우자  

### 즉, 그다음날 장이 안열리는 금요일과 토요일, 그리고 휴장일 하루전날을 빼야한다.

#### 2021년 1월 1일부터 7월 15일까지의 시계열 데이터를 생성하고 여기에 휴장된 날들을 빼보기로 하자

[2021년 휴장일](https://kakao777.tistory.com/4390)

```python
start_date = pd.to_datetime('2021-01-01') ## 시작 날짜
end_date = pd.to_datetime('2021-07-15') ## 마지막 날짜
total_dates = pd.date_range(start_date,end_date,freq='D')

weekends = []  # 금,토 빠진 리스트
holidays = []  # 주말이 아닌 연휴로 인한 휴장일을 담아둘 리스트

sub_dates = [] #최종 휴장된 날들이 빠진 리스트
```

```python
# 금,토를 빼자
for day in total_dates:
    if not (day.dayofweek == 4) and not (day.dayofweek == 5):
        weekends.append(day)
        
# 2021년 1월 1일부터 7월 15일까지 연휴로 인해 휴장된 날들은 1/1, 2/11, 2/12, 3/1, 5/5, 5/19이므로 이들의 하루 전날을 빼야한다
# 참고로 올해 첫 개장이 1월 2일까지의 날짜는 필요가 없다
for day in ['2021-01-01', '2021-01-02', '2021-02-10', 
            '2021-02-11', '2021-02-28', '2021-05-04', '2021-05-18']:
    holidays.append(datetime.strptime(day, '%Y-%m-%d'))
    

# 주말 전날을 빼준 리스트에서 공휴일 전날들을 빼자
for day in weekends:
    if day not in holidays:
        sub_dates.append(day)
```

### 이제 해당되는 날짜의 기사제목 데이터를 데이터프레임에서 빼자

```python
data_title = data_title[data_title['날짜'].isin(sub_dates)]  # sub_dates에 해당되는 것만을 남기자
data_title
```

### 합칠때 주의해야 할 점: 주가변동여부와 거래량변동여부는 기사제목의 날짜의 하루 더해진 날짜의 데이터를 넣는다

#### 예: 기사제목 ~ (날짜: 2021-07-14)의 주가변동여부, 거래량변동여부는 2021-07-15의 것으로 넣는다

```python
from tqdm import tqdm  #tqdm은 for문의 진행 정도를 시각화한다

for stock in tqdm(stocks):
    dates = data_title[(data_title['주식']==stock)]['날짜']
    dates = dates.drop_duplicates()

    for date in dates:
        data_title.loc[(data_title['주식'] == stock) & (data_title['날짜'] == date),'주가변동여부'] = data_stock[(data_stock['주식'] == stock) & (data_stock['날짜'] == date + timedelta(days=1))]['주가변동여부'].values[0]
        data_title.loc[(data_title['주식'] == stock) & (data_title['날짜'] == date),'거래량변동여부'] = data_stock[(data_stock['주식'] == stock) & (data_stock['날짜'] == date + timedelta(days=1))]['거래량변동여부'].values[0]

                                       # + timedelta(days=1)로 시간 타입 데이터에 1일을 추가한다
```

#### 날짜는 더이상 필요없으므로 뺀다

```python
data_final = data_title.drop('날짜', axis=1)
```

주가변동여부는 전일가와 비교해서 같거나 작으면 0, 크면 1의 값을 갖는다. 주가변동여부로 전일대비 주가 변동의 폭은 알 수 없다.  
거래량변동여부도 마찬가지로 전일대비 같거나 작으면 0, 크면 1이고 변동의 폭은 알 수 없다.
